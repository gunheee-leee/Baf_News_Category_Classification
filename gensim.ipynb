```python
import pandas as pd
```

```python
all_data=pd.read_csv("C:/Users/82102/Desktop/프로젝트/비어플 뉴스분류/basic_all_data.csv")
all_data
```

```python
import re

for i in range(len(all_data['text'])):
    #all_data['text'][i]=re.sub(r'[\n\t\\]','', all_data['text'][i])
    all_data['text'][i]=re.sub(r'\n', ' ', all_data['text'][i])
    all_data['text'][i]=re.sub(r'\t', ' ', all_data['text'][i])
    all_data['text'][i]=re.sub(r"[^가-힣0-9'\". ]", "", all_data['text'][i])
```

```python
all_data.drop(columns=['Unnamed: 0'], inplace=True)
all_data
```

```python
all_data.reset_index(drop=True, inplace=True)
```

```python
all_data['length'] = all_data['text'].apply(lambda x: len(x))
all_data
```

```python
all_data.to_csv('test1.csv', index=True)
```
